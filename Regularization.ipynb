```python
from sklearn.datasets import make_moons

x, y = make_moons(100, noise=.25)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x[ : , 0],x[ : , 1], c=y)
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.regularizers import l1

model1 = Sequential()
model1.add(Dense(128, input_dim = 2, activation = 'relu'))
model1.add(Dense(128, activation = 'relu'))
model1.add(Dense(1, activation = 'sigmoid'))

model2 = Sequential()
model2.add(Dense(128, input_dim = 2, activation = 'relu', kernel_regularizer = l1(.001)))
model2.add(Dense(128, activation = 'relu', kernel_regularizer = l1(.001)))
model2.add(Dense(1, activation = 'sigmoid'))

model1.summary()
model2.summary()
```

```python
model1.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model2.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
hist1 = model1.fit(x, y, epochs=100, validation_split=0.2)
hist2 = model2.fit(x, y, epochs=100, validation_split=0.2)
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x, y, clf=model1)
plt.legend()
plt.show()

plot_decision_regions(x, y, clf=model2)
plt.legend()
plt.show()
```

```python
#Plotting
plt.plot(hist1.history['accuracy'], label='Model 1 Train')
plt.plot(hist1.history['val_accuracy'], label='Model 1 Test')
plt.plot(hist2.history['accuracy'], label='Model 2 Train')
plt.plot(hist2.history['val_accuracy'], label='Model 2 Test')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(hist1.history['loss'], label='Model 1 Train')
plt.plot(hist1.history['val_loss'], label='Model 1 Test')
plt.plot(hist2.history['loss'], label='Model 2 Train')
plt.plot(hist2.history['val_loss'], label='Model 2 Test')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
print('Accuracy for model 1:', hist1.history['accuracy'][-1])
print('Accuracy for model 2:', hist2.history['accuracy'][-1])
```

```python
import pandas as pd
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
df = df.dropna()
```

```python
legit = df[df['Class']==0]
fraud = df[df['Class']==1]
print(legit.shape)
print(fraud.shape)
```

```python
#undersample
new_legit = legit.sample(n=264)
new_legit.shape
```

```python
new_df = pd.concat([new_legit,fraud])
new_df['Class'].value_counts()
```

```python
x= new_df.drop('Class', axis=1)
y= new_df['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
pred = dtc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(accuracy_score(pred, y_test))
print(classification_report(pred, y_test))
```
